```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

```python
np.random.seed(42)
x=[111,33,84,45,67,64,79,86,17,90,53,46,29,100,30]
y=[]
for i in x:
    y.append(i*2)
len(x)
```

```python
train_input=np.array(x[0:11]).reshape(11,1)
train_output=np.array(y[0:11]).reshape(11,1)
test_input=np.array(x[11:]).reshape(len(x)-11,1)
test_output=np.array(y[11:]).reshape(len(x)-11,1)
```

f(x)=mx+c


```python
def forward_propogation(train_input,parameters):
    m=parameters['m']
    c=parameters['c']
    prediction=np.multiply(m,train_input)+c
    return prediction
```

#cost function
1/2n(y-f(x))^2

```python
def cost_function(prediction,train_output):
    cost=np.mean((train_output-prediction)**2)*0.5
    return cost
```

gradient descent 

df=f(x)-y
dm=df*x
dc=df

```python
def backward_propogation(train_input,train_output,prediction):
    derivative=dict()
    df=prediction-train_output
    dm=np.mean(np.multiply(df,train_input))
    dc=np.mean(df)
    derivative['dm']=dm
    derivative['dc']=dc
    return derivative
```

```python
def update_par(parameters,derivative,learning_rate):
    parameters['m']=parameters['m']-learning_rate*derivative['dm']
    parameters['c']=parameters['c']-learning_rate*derivative['dc']
    return parameters
```

```python
def train(train_input,train_output,learning_rate,iters):
    parameters=dict()
    parameters['m']=np.random.uniform(0,1)
    parameters['c']=np.random.uniform(0,1)
    plt.figure()
    loss=list()
    for i in range(iters):
        prediction=forward_propogation(train_input,parameters)
        cost=cost_function(prediction,train_output)
        loss.append(cost)
        fig,ax=plt.subplots()
        print(f'iteration {i+1}')
        ax.plot(train_input,train_output,label="ori")
        ax.plot(train_input,prediction,label="tra")
        legend=ax.legend()

        
        
        print(parameters)
        plt.show()
        derivatives=backward_propogation(train_input,train_output,prediction)
        parameters=update_par(parameters,derivatives,learning_rate)
    return parameters,loss

```

```python
parameters , loss =train(train_input,train_output,0.0001,10)
```
