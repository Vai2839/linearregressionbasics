```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Dataset.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
x=data.iloc[:, :-1]
y=data.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
x_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
cost=mean_squared_error(x_pred,y_test)
cost
```

```python

plt.plot(x_test,x_pred,'*',label="predic",color="red")
plt.plot(x_test,y_test,'+',label="actual",color="green")

plt.show()
```
