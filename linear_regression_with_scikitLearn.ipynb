```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

```python
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)
x=[111,33,84,45,67,64,79,86,17,90,53,46,29,100,30]
y=[]
for i in x:
    y.append(i*2)
len(x)
train_input=np.array(x[0:11]).reshape(11,1)
train_output=np.array(y[0:11]).reshape(11,1)
test_input=np.array(x[11:]).reshape(len(x)-11,1)
test_output=np.array(y[11:]).reshape(len(x)-11,1)
```

```python
lr_model=LinearRegression()
```

```python
lr_model.fit(train_input,train_output)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

```python
test_pred=lr_model.predict(test_input)
```

```python
plt.plot(test_input,test_output,color='red')
plt.plot(test_input,test_pred,color='purple')
plt.xlabel='input'
plt.ylabel='output'
plt.title="prediction"
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
cost=mean_squared_error(test_output,test_pred)
cost
```
